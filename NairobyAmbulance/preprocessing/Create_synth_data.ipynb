```python
import numpy as np
import pandas as pd
import geopandas as gpd
```

```python
data_merged = pd.read_csv('../Data/data_merged_limited.csv')
```

```python
data_merged
```

```python
data_merged = data_merged.drop(['longitude', 'latitude'], axis=1)
```

```python
data_merged['hour'] = pd.to_datetime(data_merged['time']).dt.hour
data_merged
```

```python
data_merged['hour'].unique()
```

```python
def transform_hour(hour):
    if hour >= 0 and hour <3: 
        return 1
    elif hour >= 3 and hour <6: 
        return 2
    elif hour >= 6 and hour <9: 
        return 3
    elif hour >= 9 and hour <12: 
        return 4
    elif hour >= 12 and hour <15: 
        return 5
    elif hour >= 15 and hour <18: 
        return 6
    elif hour >= 18 and hour <21: 
        return 7
    elif hour >= 21: 
        return 8
data_merged['hour'] = data_merged['hour'].apply(transform_hour)
data_merged['hour'].unique()
```

```python
data_grouped_per_hour = data_merged.groupby([data_merged['hour'], data_merged['date'], data_merged['segment_id']])
```

```python
data_grouped_per_hour
```

```python
df_group = data_grouped_per_hour.count().reset_index()  
```

```python
df_group
```

```python
df_group.to_csv('../Data/data_merged_time_int.csv', index=False)
```
