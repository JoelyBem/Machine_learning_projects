### Gradient descent for distance minimisation:
- Limite the dataset by removing outliers [../data/data_train_limited.csv]
- Initialize random 8*6=48 locations for 6 ambulances and 8 hour interval
- For each crash:
    - Consider hour interval
    - Caclulate the distance between crash and all ambulance
    - Taking the minimum distance
- Loss is equal to the mean of distance of all crash and ambulance
- We consider only one model for all crash 
- Consider longitude, latitude and hour for crash
- We have 8 models: one model for one hour interval

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import numpy as np
from sklearn.cluster import KMeans
import random
import datetime
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('../data/data_train_limited.csv',parse_dates=['datetime'])
print(df.shape)
df.head()
```

```python
# View the submission format
ss = pd.read_csv('../Data/SampleSubmission.csv', parse_dates=['date'])
ss.head()
```

```python
amb_loc_hour = tf.Variable(tf.random.normal([8, 6, 2], mean=0, stddev=1)) * 0.04 + [-1.27, 36.85]
```

```python
def loss_fn(X, amb_loc_hour, key_hour):
    dist_split = X - amb_loc_hour[key_hour]
    dist = tf.sqrt(tf.reduce_sum(tf.square(dist_split), axis=2))
    min_distance = tf.reduce_min(dist, axis=1)
    cost = tf.reduce_mean(min_distance)
    return cost
```

```python
X = tf.placeholder(tf.float32, [None, 1, 2])
key_hour = tf.placeholder(tf.int32, [])
cost = loss_fn(X, amb_loc_hour, key_hour)
learning_rate = tf.placeholder(tf.float32, [])

optimizer_hour = {'[0,3]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[3,6]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[6,9]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[9,12]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[12,15]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[15,18]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[18,21]': tf.train.AdamOptimizer(learning_rate).minimize(cost),
                 '[21,24]': tf.train.AdamOptimizer(learning_rate).minimize(cost)} 
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
cost_history = {'[0,3]': [], '[3,6]': [], '[6,9]': [], '[9,12]': [],
                    '[12,15]': [], '[15,18]': [], '[18,21]': [], '[21,24]': []}
training_epochs = 1000
mini_batch_size = 50
learning_r = 0.01

for epoch in range(training_epochs + 1):
    for hour in range(0,22,3):
        df_hour = df[(df['datetime'].dt.hour>=hour) & (df['datetime'].dt.hour<=hour)]
        key = '[' + str(hour) + ',' + str(hour+3) + ']'
        X_train_batch = df_hour[['latitude', 'longitude']].values.reshape([-1, 1, 2])
        sess.run(optimizer_hour[key], feed_dict = {X: X_train_batch, learning_rate: learning_r, key_hour: int(hour/3)})
        cost_ = sess.run(cost, feed_dict = {X: X_train_batch, key_hour: int(hour/3)})
        cost_history[key].append(cost_)
    if epoch % 100 == 0:
        print('In epoch: ', epoch, 'cost = ', cost_)
```

```python
amb_loc_np = amb_loc_hour.eval(session = sess)
```
