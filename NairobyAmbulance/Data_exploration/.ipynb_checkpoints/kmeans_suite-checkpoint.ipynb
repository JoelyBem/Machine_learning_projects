### Trying another approach for clustering (Considering only 3 cluster):
- Choose different kmeans according to time interval [Exploration_data_all Describe location of crash per hour interval]:
    - Between 00h and 06h
    - Between 06h and 09h
    - Between 09h and 00h

```python
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import numpy as np
from sklearn.cluster import KMeans
import random
import datetime
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('../Data/Train_mod.csv',parse_dates=['datetime'])
print(df.shape)
df.head()
```

```python
# View the submission format
ss = pd.read_csv('../Data/SampleSubmission.csv', parse_dates=['date'])
ss.head()
```

```python
def flaten(centers_list):
    """flatening the list of tuples [(x0,y0),(x1,y1),(x2,y2),(x3,y3)] to [x0,y0,x1,y1,x2,y2,x3,y3]"""
    merged = list(zip(centers_list[:,0],centers_list[:,1]))
    return [c for couple in merged for c in couple ]
```

### Define 3 different model

```python
df_0_6 = df.loc[(df['datetime'].dt.time >= datetime.time(0, 0, 0)) & (df['datetime'].dt.time < datetime.time(6, 0, 0))]
df_6_9 = df.loc[(df['datetime'].dt.time >= datetime.time(6, 0, 0)) & (df['datetime'].dt.time < datetime.time(9, 0, 0))]
df_9_0 = df.loc[(df['datetime'].dt.time >= datetime.time(9, 0, 0))]
```

```python
kmeans_0_6 = KMeans(n_clusters=6, random_state=0).fit(df_0_6[['latitude','longitude']])
kmeans_6_9 = KMeans(n_clusters=6, random_state=0).fit(df_6_9[['latitude','longitude']])
kmeans_9_0 = KMeans(n_clusters=6, random_state=0).fit(df_9_0[['latitude','longitude']])
```

```python
fig, axes = plt.subplots(8, 1, figsize=(10, 50))
m = 0
for i in range(0,22,3):
    if i >= 0 and i < 6:
        d = df[(df['datetime'].dt.hour>=i) & (df['datetime'].dt.hour<=i+3)]
        t = flaten(kmeans_0_6.cluster_centers_)
        ss.loc[(ss['date'].dt.hour == i), ['A0_Latitude','A0_Longitude','A1_Latitude','A1_Longitude','A2_Latitude','A2_Longitude','A3_Latitude','A3_Longitude','A4_Latitude','A4_Longitude','A5_Latitude','A5_Longitude']] = t
        axes[m].scatter(x = d['latitude'],y=d['longitude'],s=10, label='Craches locations')
        axes[m].set_title(' Interval hour: ' + str(i)+':00h' + ' ' + str(i+3) + ':00h')
        axes[m].scatter(x=kmeans_0_6.cluster_centers_[:,0], y=kmeans_0_6.cluster_centers_[:,1],marker="x", color='r',label='Ambulances locations at '+str(i)+':00h',s=40)
        axes[m].legend()
    elif i>=6 and i<9:
        d = df[(df['datetime'].dt.hour>=i) & (df['datetime'].dt.hour<=i+3)]
        t = flaten(kmeans_6_9.cluster_centers_)
        ss.loc[(ss['date'].dt.hour == i), ['A0_Latitude','A0_Longitude','A1_Latitude','A1_Longitude','A2_Latitude','A2_Longitude','A3_Latitude','A3_Longitude','A4_Latitude','A4_Longitude','A5_Latitude','A5_Longitude']] = t
        axes[m].scatter(x = d['latitude'],y=d['longitude'],s=10, label='Craches locations')
        axes[m].set_title(' Interval hour: ' + str(i)+':00h' + ' ' + str(i+3) + ':00h')
        axes[m].scatter(x=kmeans_6_9.cluster_centers_[:,0], y=kmeans_6_9.cluster_centers_[:,1],marker="x", color='r',label='Ambulances locations at '+str(i)+':00h',s=40)
        axes[m].legend()
    else:
        d = df[(df['datetime'].dt.hour>=i) & (df['datetime'].dt.hour<=i+3)]
        t = flaten(kmeans_9_0.cluster_centers_)
        ss.loc[(ss['date'].dt.hour == i), ['A0_Latitude','A0_Longitude','A1_Latitude','A1_Longitude','A2_Latitude','A2_Longitude','A3_Latitude','A3_Longitude','A4_Latitude','A4_Longitude','A5_Latitude','A5_Longitude']] = t
        axes[m].scatter(x = d['latitude'],y=d['longitude'],s=10, label='Craches locations')
        axes[m].set_title(' Interval hour: ' + str(i)+':00h' + ' ' + str(i+3) + ':00h')
        axes[m].scatter(x=kmeans_9_0.cluster_centers_[:,0], y=kmeans_9_0.cluster_centers_[:,1],marker="x", color='r',label='Ambulances locations at '+str(i)+':00h',s=40)
        axes[m].legend()
    m += 1
plt.show()
```

```python
ss.to_csv('kmeans_3h_3_model.csv')
```
