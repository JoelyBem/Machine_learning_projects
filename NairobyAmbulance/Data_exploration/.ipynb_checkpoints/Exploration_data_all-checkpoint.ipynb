```python
df_count_crash_date = pd.DataFrame(columns=['Date', 'Count'])
list_date = []
list_count = []
for date in data_train['date'].unique():
    list_date.append(date)
    list_count.append(len(data_train.loc[data_train['date'] == date]))
df_count_crash_date['Date'] = list_date
df_count_crash_date['Count'] = list_count
```

```python
data_train['time'] = data_train['datetime'].dt.time
```

### Counting crash per hour

```python
df_crash_per_hour = pd.DataFrame(columns=["Hour Interval", "Count"])
hour_prec = 0
list_hour = []
list_count = []
for hour in range(1, 25):
    data_for_hour = data_train.loc[(datetime.time(hour_prec%24, 0, 0) < data_train['time']) & (data_train['time'] <= datetime.time(hour%24, 0, 0))]
    list_hour.append("[" + str(hour_prec) + "," + str(hour) + "]")
    list_count.append(len(data_for_hour))
    hour_prec = hour
df_crash_per_hour["Hour Interval"] = list_hour
df_crash_per_hour["Count"] = list_count
```

```python
df_crash_per_hour
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x=df_crash_per_hour['Hour Interval'], y=df_crash_per_hour['Count'], sort=False)
```

### Describe location of crash per hour interval

```python
img_list = []
hour_prec = 0
dict_crash_per_hour = {}
f, axes = plt.subplots(12, 2, figsize=(20, 50), sharex=True, sharey=True)
i, j = 0, 0
for hour in range(1, 25):
    data_for_hour = data_train.loc[(datetime.time(hour_prec%24, 0, 0) < data_train['time']) & (data_train['time'] <= datetime.time(hour%24, 0, 0))]
    sns.scatterplot(data=data_for_hour, x='longitude', y='latitude', color='black', ax=axes[i, j]).set_title(
        "Between: " + str(datetime.time(hour_prec%24, 0, 0)) + " and " + str(datetime.time(hour%24, 0, 0)) + "\n Crash count: "
    + str(len(data_for_hour)))
    hour_prec = hour
    j += 1
    if j==2:
        i+=1
        j=0
f.tight_layout(pad=3.0)
```

### Conclusion:
- La plupart des accidents sont identifis avec une longitude infrieur  37
- Les accidents les plus nombreuses sont entre 06:00 et 23:00 [Suprieur  200]
- Pour chaque intervalle, il existe une location spcifique o les accidents se contentrent

### Hypothesis:
- Number of crash expands with the number of trafics
- The most crash is identify during the rush hour

## Exploration weather.csv

### Plot mean of precipitable_water per month

```python
data_weather["year_month"] = data_weather["Date"].dt.strftime('%Y-%m')
```

```python
data_weather = data_weather.loc[data_weather["Date"].dt.date <= datetime.date(2019, 6, 30)]
```

```python
df_prec_water_mean = pd.DataFrame(columns=['Month Year', 'Mean precipitable water'])
list_month_year = []
list_mean_prec_water = []
for year_month in data_weather['year_month'].unique():
    list_month_year.append(year_month)
    list_mean_prec_water.append(data_weather.loc[data_weather['year_month'] == year_month]['precipitable_water_entire_atmosphere'].mean())
df_prec_water_mean['Month Year'] = list_month_year
df_prec_water_mean['Mean precipitable water'] = list_mean_prec_water
```

## Exploration of Segment_info.csv

```python
# Load the survey data
road_surveys = pd.read_csv('../Data/Segment_info.csv')
road_surveys.head()
```

```python
# Load the map
road_segment_locs = gpd.read_file('../Data/segments_geometry.geojson')
road_segment_locs.head()
```

#### Ploting segments with crashes coordinates (All accident)

```python
# Because this is a geodataframe, we can plot it
f, axes = plt.subplots(1, 1, figsize=(20, 20), sharex=False)
road_segment_locs.plot(ax=axes)
sns.scatterplot(data=data_train, x='longitude', y='latitude', color='black', ax=axes, s=10)
```

#### Ploting segments with crashes coordinates (Limit by the segments localisation)

```python
# Because this is a geodataframe, we can plot it
f, axes = plt.subplots(1, 1, figsize=(20, 20), sharex=False)
axes.set_xlim([36.7, 37.1])
axes.set_ylim([-1.375 , -1.175])
road_segment_locs.plot(ax=axes, figsize=(5,5))
frequency = round(len(data_train.loc[(data_train["longitude"] >= 36.1) & 
                               (data_train["longitude"] <= 37.1) & (data_train["latitude"] >= -1.375) 
                               & (data_train["latitude"] <= -1.175)])/len(data_train) * 100, 2)
sns.scatterplot(data=data_train, x='longitude', y='latitude', color='black', ax=axes, s=10).set_title(
    "Number of accident in the interval: " + str(frequency))
```

### Plot accident and segment per hour interval

```python
img_list = []
hour_prec = 0
dict_crash_per_hour = {}
f, axes = plt.subplots(12, 2, figsize=(20, 50), sharex=True, sharey=True)
i, j = 0, 0
for hour in range(1, 25):
    axes[i, j].set_xlim([36.7, 37.1])
    axes[i, j].set_ylim([-1.375 , -1.175])
    data_for_hour = data_train.loc[(datetime.time(hour_prec%24, 0, 0) < data_train['time']) & (data_train['time'] <= datetime.time(hour%24, 0, 0))]
    road_segment_locs.plot(ax=axes[i, j], figsize=(5,5))
    sns.scatterplot(data=data_for_hour, x='longitude', y='latitude', color='black', ax=axes[i, j], s=15).set_title(
        "Between: " + str(datetime.time(hour_prec%24, 0, 0)) + " and " + str(datetime.time(hour%24, 0, 0)) + "\n Crash count: "
    + str(len(data_for_hour)))
    hour_prec = hour
    j += 1
    if j==2:
        i+=1
        j=0
f.tight_layout(pad=3.0)
```

### Plot accident and segment per day of week

```python
dict_crash_per_day = {}
f, axes = plt.subplots(4, 2, figsize=(20, 20), sharex=True, sharey=True)
i, j = 0, 0
for day in data_train["day_of_week"].unique():
    axes[i, j].set_xlim([36.7, 37.1])
    axes[i, j].set_ylim([-1.375 , -1.175])
    road_segment_locs.plot(ax=axes[i, j], figsize=(5,5))
    data_for_day = data_train.loc[data_train["day_of_week"] == day]
    freq_in_interval = round(len(data_for_day.loc[(data_for_day["longitude"] >= 36.1) & 
                               (data_for_day["longitude"] <= 37.1) & (data_for_day["latitude"] >= -1.375) 
                               & (data_for_day["latitude"] <= -1.175)])/len(data_for_day)*100, 2)
    sns.scatterplot(data=data_for_day, x='longitude', y='latitude', color='black', ax=axes[i, j], s=15).set_title(str(day) + ": "
                                                      + str(len(data_for_day)) + "\n" + str(freq_in_interval))
    j += 1
    if j==2:
        i+=1
        j=0
f.tight_layout(pad=3.0)
```
